# Project: Hangman

Game Overview
The player must guess a hidden word one letter at a time before running out of lives.

Minimum Requirements
Data Structures
Your game must include:

A list containing at least 10 possible words.
A game state dictionary to track progress.
Gameplay
Randomly select a word at the start of the game.
Allow the player to guess one letter at a time.
Reveal correctly guessed letters.
Keep hidden letters concealed.
Reduce the player's lives after incorrect guesses.
Game State - Track:
-Secret word
-Guessed letters
-Remaining lives

Win Condition
The player successfully reveals the entire word.

Lose Condition
The player runs out of lives.

Additional Requirements
Display the current word progress after every guess.
Prevent duplicate guesses from affecting the game.
Validate user input.


## Pseudocode

## Initial Values:

```python
from math import ceil

#List of words
thesaurus = [
    "adventure",
    "galaxy",
    "strategy",
    "penguin",
    "language",
    "volcano",
    "hospital",
    "mystery",
    "resource",
    "workshop"
]

#Dictionary to track the status of the game as it proceeds
game_state = {
    'secret_word' : '',
    'guessed_letters' : [],
    'remaining_lives' : '',
    'hasWon' : False,
    'display_list' : []
}

#Number of lives : Established in relation to the average lenght of the words in the thesaurus
allowed_lives = sum(len(word) for word in thesaurus) / len(thesaurus)
allowed_lives = ceil(allowed_lives)
game_state['remaining_lives'] = allowed_lives

print(game_state)
```

## Select the secret word to use

```python
import random

#We get a random number between 0 and 9 to be used as the index of the selected word forom the thesaurus list.
index_secret_word = random.randint(0, 9)

#Update theselected word in the game state
game_state['secret_word'] = thesaurus[index_secret_word]

game_state['display_list'] = ["_" for letter in game_state['secret_word']]

print(game_state)
```

## Logic

```python
## Ask for input
```

```python
# action: ask for input
def askForInput():
    input_letter= input("Please give me a guess - Just include a single letter at a time- ").strip()
    #print(len(input_to_check))

    
    while not isValidInput(input_letter):
        print("Please enter a single letter.")
        input_letter = input("Enter a letter: ").strip()
    
    print(f'Your guess was {input_letter.upper()}')
    return input_letter.lower()

#askForInput()
```

```python
def isValidInput(input_text):
    return len(input_text) == 1 and input_text.isalpha()
#isValidInput('j')
```

```python
## Y. Add letters
```

```python
#Y. add letter to list of guessed letters
def addLetters(letter):
    game_state['guessed_letters'].append(letter)

#addLetters(input_letter)
#print(game_state)
#print(type(game_state['guessed_letters']))
```

```python
## X. Reduce lives
```

```python
def reduceLives():
    game_state['remaining_lives'] = game_state['remaining_lives'] -1

#reduceLives()
#print(game_state)
```

```python
## Z. Update display
```

```python
def update_display():
    for index in range(len(game_state['secret_word'])):
        if game_state['secret_word'][index] in game_state['guessed_letters']:
            #substitute
            game_state['display_list'][index] = game_state['secret_word'][index]
        else:
            game_state['display_list'][index] = '_'
    display_string = " ".join(game_state['display_list'])
    print(f' \n {display_string} \n')
    

update_display()
#print(f'The secret word {game_state['secret_word']} has {len(thesaurus[index_secret_word])} letters')
#print(display_list)
print(game_state)
```

```python
def checkIfWon():
    # How many unique letters are in the word? Using a set because it has a list of unique(not repeated) values.
    unique_letters = set(game_state['secret_word'])
    
    # if all unique letters are in a set of the guessed letters.
    if unique_letters <= set(game_state['guessed_letters']):
        game_state['hasWon'] = True

#checkIfWon()
#print(game_state)
```

```python

#PLAY
def play():
    print(f'\n Remaining lives: {game_state['remaining_lives']}')
    update_display()
    input_letter = askForInput()
    
    # option i: Ya preguntado
    if input_letter in set(game_state['guessed_letters']):
        print('You had already used that letter.')
        
    #option ii: nuevo, correcto
    elif input_letter in set(game_state['secret_word']):
        addLetters(input_letter)
        print('That was a correct guess!')
        checkIfWon()
        #reduceLives()

    #option iii: nuevo, incorrecto
    elif input_letter not in set(game_state['secret_word']):
        addLetters(input_letter)
        print('That letter is not in the word :( Keep trying!')
        reduceLives()

    
    #print(game_state)
    

#play()
```

## Ready to play the game!!

```python
#WHILE LOOP

print("Welcome to hangman! Please guess one letter at a time")

while game_state['remaining_lives']>0 and game_state['hasWon'] == False:
    play()
else:
    if game_state['remaining_lives'] == 0:
        print('Sorry you have exceeded the number of attempts')
    elif game_state['hasWon'] == True:
        update_display()
        print ('Congratulations, you won!')
        

print('End of Game \n')

#print(game_state)
```
